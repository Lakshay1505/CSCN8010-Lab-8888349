```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("/Users/lakshay/Desktop/ML_Course/FINAL_GIT/CSCN8010-Lab-8888349/Data/housing.csv")

# Define unique conditions
unique_condition = (df['population'] > df['population'].mean()) & (df['median_income'] > df['median_income'].mean())

# Create scatter plot for median_income vs. median_house_value
plt.figure(figsize=(10, 6))

# Plot all data points
plt.scatter(df['median_income'], df['median_house_value'], alpha=0.5, label='All Data')

# Highlight unique cases with a different marker or color
plt.scatter(df.loc[unique_condition, 'median_income'], df.loc[unique_condition, 'median_house_value'], color='red', marker='X', label='Unique Cases')

# Set plot labels and title
plt.xlabel('Median Income')
plt.ylabel('Median House Value')
plt.title('Scatter Plot: Median Income vs. Median House Value')

# Show legend
plt.legend()

# Show the plot
plt.show()
```
